```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers = ['PG', 'MSFT', 'F', 'GE']
mydata = pd.DataFrame()
for t in tickers:
    mydata[t] = wb.DataReader(t, data_source='yahoo', start='1995-01-01')['Adj Close']
```

```python
returns = (mydata / mydata.shift(1)) -1 
returns.head()
```

```python
weighs = np.array([0.25,0.25,0.25,0.25])
```

```python
np.dot(returns, weights)
```

```python
annual_returns = returns.mean() * 250
annual_returns
```

```python
pfolio_1 = str(round(np.dot(annual_returns, weighs), 5) * 100) +'%'
print(pfolio_1)
```

```python
weights_2 = np.array([0.4,0.4,0.15,0.05])
```

```python
pfolio_2 = str(round(np.dot(annual_returns, weights_2), 5) * 100) + '%'
print(pfolio_2)
```
