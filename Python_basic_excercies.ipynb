```python
def genStr(str1, str2):
    first_char = str1[:1] + str2[:1]
    middle_char = str1[int(len(str1) /2) : int(len(str1) / 2) + 1] + str2[int(len(str2) / 2) : int(len(str2) / 2) + 1]   
    last_char = str1[len(str1) -1] + str2[len(str2) -1]
    return first_char + middle_char + last_char
    
    
s1 = "America"
s2 = "Japan"
s3 = "Xianhina"
genStr(s1,s3)
```

```python
def sortStr(str):
    for i in range(len(str)):
        if ( str[i].isupper()):
            str = str[: i] + str[i+1: len(str)] + str[i]
    return str
               
print(sortStr("PyNaTive"))  
print(sortStr("ChineseJapanAmerica"))
```

```python
def countAll(str):
    up_num = 0
    low_num = 0
    digit_num = 0
    symbol_num = 0
    for i in range(len(str)):
        if (str[i].isnumeric()): 
            digit_num += 1
        elif(str[i].isalpha()):
            if(str[i].isupper()):
                up_num += 1
            else: 
                low_num += 1
        else: 
            symbol_num += 1
           
     
    print("char = ", up_num + low_num)
    print("uppercase = ", up_num)
    print("lowercase = ", low_num)
    print("digit = ", digit_num)
    print("special symbol = ", symbol_num)
            
countAll("P@#yn26at^&i5ve")  
```

```python
# generate password, with at least 2 uppercase letters, 1 digit, 1 special symbol
import random
import string

def passwordGen(pwd_length):
    randomSource = string.ascii_letters + string.digits + string.punctuation
    pwdStr = random.sample(randomSource, pwd_length - 4)
    pwdStr += random.sample(string.ascii_uppercase,2)
    pwdStr += random.choice(string.digits)
    pwdStr += random.choice(string.punctuation)
    
    #print(pwdStr)
    random.shuffle(pwdStr)
    password = ''.join(pwdStr)
    #print(pwdStr)
    return password

print(passwordGen(10))
```

```python
import datetime
import calendar

startDate = (1900,4, 12)
endDate = (2020,4,12)
    
        
def genRandomDate(startDate, endDate):
    bigMon = [1,3,5,7,8,10,12]
    genYear = random.randint(startDate[0], endDate[0])
    genMon = random.randint(1,12)
        
    if (genMon == 2):  
        if (calendar.isleap(genYear)):
                genDay = random.randint(1,28)
        else:
                genDay = random.randint(1,29)
    elif(genMon in bigMon):
        genDay = random.randint(1,31)
    else:
        genDay = random.randint(1,30)
    return datetime.date(genYear,genMon,genDay)

print(genRandomDate(startDate,endDate))
#print(calendar.isleap(2000))
    
```

```python
import random
import time

def genRandDate(startDate, endDate):
    dateFormat = '%m/%d/%Y'
    
    startTime = time.mktime(time.strptime(startDate, dateFormat))
    endTime = time.mktime(time.strptime(endDate, dateFormat))
    
    randDateTime = (endTime - startTime) * random.random() + startTime
    randDate = time.strftime(dateFormat, time.localtime(randDateTime))
    
    return randDate

print(genRandDate('1/1/1900', '12/31/2020'))
```
